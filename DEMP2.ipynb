```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
visit=vt.replace(np.nan,0)
```

```python
sns.heatmap(visit.corr(), vmin=-1, vmax=1, cmap="Blues") 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X = visit[['Business Page Visit','Buy Page Visit','Choose Plan Visit','Completed Contact Us Form','Contact Page Visit','Plans Page Visit','Pricing Page Visit','Resource Page Visit','Try Page Visit','Webinar Signup Page Visit']]
y = visit['Conversion']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,y_train)
y_pred=logistic_regression.predict(X_test)
```

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
plt.show()
```

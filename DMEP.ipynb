```python
# import packages
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#replace null with 0 in the data frame
visit=vt.replace(np.nan,0)
```

```python
#check new data frame
visit.head()
```

```python
#drop columns with no value
visit=visit.drop(['Pricing Page Visit'], axis=1)
```

```python
#check data types
visit.info()
```

```python
#check data statistics
visit.describe()
```

```python
#check correlations
visit.corr()
```

```python
#plot data
sns.heatmap(visit.corr(), vmin=-1, vmax=1,  cmap="coolwarm_r") 
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
```

```python
#import packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#set variables
X = visit[['Business Page Visit','Buy Page Visit','Choose Plan Visit','Completed Contact Us Form','Contact Page Visit','Plans Page Visit','Resource Page Visit','Try Page Visit','Webinar Signup Page Visit']]
y = visit['Conversion']
```

```python
#split training and testing data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#run regression
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,y_train)
y_pred=logistic_regression.predict(X_test)
```

```python
#check acturacy on actual vs predicted
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
#check accuracy
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
plt.show()
```
